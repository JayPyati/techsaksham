```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image using OpenCV
image = cv2.imread("cat.jpeg")  # Reads in BGR format

# Convert BGR to RGB (Matplotlib expects RGB format)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image_rgb)
plt.axis("off")  # Hide axes
plt.show()
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap="gray")
plt.axis("off")
plt.show()
```

```python
resized_image = cv2.resize(image, (300, 300))
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
print("Image Shape:", image.shape)  # (height, width, channels)
print("Data Type:", image.dtype)
print("First Pixel (BGR):", image[0, 0])  # Accessing pixel values
```

```python
blurred = cv2.GaussianBlur(image, (15, 15), 0)
plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
edges = cv2.Canny(image, 100, 200)  # Min and Max threshold
plt.imshow(edges, cmap="gray")
plt.axis("off")
plt.show()
```

```python
import pandas as pd

# Convert image to a 2D NumPy array (flattening)
gray_flat = gray_image.flatten()

# Create a Pandas DataFrame
df = pd.DataFrame(gray_flat, columns=["Pixel Intensity"])
print(df.head())  # Display first few pixel values
```

```python
plt.hist(gray_image.ravel(), bins=256, range=[0, 256], color="black", alpha=0.7)
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.title("Grayscale Histogram")
plt.show()
```

```python
import cv2
import numpy as np
import pyautogui

# Get screen size
screen_size = pyautogui.size()
fps = 20.0  # Frames per second

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*"XVID")
out = cv2.VideoWriter("screen_record.avi", fourcc, fps, screen_size)

print("Recording... Press Ctrl+C to stop.")

try:
    while True:
        # Capture the screen
        img = pyautogui.screenshot()
        frame = np.array(img)  # Convert to NumPy array
        frame = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)  # Convert RGB to BGR

        # Write to video file
        out.write(frame)

        # Show preview (optional)
        cv2.imshow("Recording", frame)
        if cv2.waitKey(1) == ord("q"):  # Press 'q' to stop recording
            break

except KeyboardInterrupt:
    print("Recording stopped.")

finally:
    out.release()
    cv2.destroyAllWindows()
    print("Video saved as 'screen_record.avi'")
```
